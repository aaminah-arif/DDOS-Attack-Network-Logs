## Data Preprocessing
This Python file generates a CSV file containing a random sample of 2 million data points from the original dataset.

We performed this step because GitHub has file size limitations. Since we were collaborating on GitHub, we couldn't upload the large dataset directly. Instead, we created a smaller sample locally and uploaded the sampled CSV file to the repository.

```python
#File to Generate a small random sample from the 2million samples for model training
from scipy.io import arff #needed to read arff files
import random
import pandas as pd
```

```python
data, meta = arff.loadarff('ddos_data/final-dataset.arff') #this is the full version of the dataset with 2 million entries
df = pd.DataFrame(data)
df
```

```python
#create a sample of 1000 random samples from the dataset
sample = df.sample(n=1000, random_state=42) #random_state is set to 42 for reproducibility
sample.head() #display the first 5 rows of the sample to verify that it was created successfully
#display the shape of the sample to verify that it has 1000 rows and 28 columns 
print(sample.shape) #should print (1000, 28)
```
